```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_excel("CACS_model_ready_imputed2.xlsx")

# 2. Î∂ÑÏÑù ÎåÄÏÉÅ Î≥ÄÏàò ÏÑ†ÌÉù (CACS Ï†úÏô∏Ìïú ÏàòÏπòÌòï Î≥ÄÏàò ÏûêÎèô Í∞êÏßÄ)
numerical_vars = df.select_dtypes(include=["float64", "int64"]).drop(columns=["id", "CACS"]).columns.tolist()

# 3. CACSÏôÄ Îã§Î•∏ Î≥ÄÏàòÎì§ Í∞ÑÏùò ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
correlations = df[numerical_vars + ["CACS"]].corr(method='pearson')["CACS"].drop("CACS").sort_values(ascending=False)

# 4. ÏãúÍ∞ÅÌôî (ÎßâÎåÄÍ∑∏ÎûòÌîÑ)
plt.figure(figsize=(8, 5))
sns.barplot(x=correlations.values, y=correlations.index, palette="viridis")
plt.title("CACSÏôÄ Îã§Î•∏ Î≥ÄÏàò Í∞Ñ Pearson ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ")
plt.xlabel("ÏÉÅÍ¥ÄÍ≥ÑÏàò (r)")
plt.axvline(0, color="gray", linestyle="--")
plt.tight_layout()
plt.show()

# 5. ÏÉÅÍ¥ÄÍ≥ÑÏàò ÏàòÏπò Ï∂úÎ†•
print("üìä CACSÏôÄ Í∞Å Î≥ÄÏàò Í∞Ñ ÏÉÅÍ¥ÄÍ≥ÑÏàò:")
print(correlations.round(3))
```
