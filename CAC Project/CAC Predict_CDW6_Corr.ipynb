```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_excel("CACS_model_ready_imputed.xlsx")

# 2. 상관관계 분석 대상 변수 선택
features = df.columns.tolist()
target = "CACS"

# 3. CACS와 다른 변수들 간의 상관계수 계산
correlations = df[features].corr(method='pearson')[target].drop(target).sort_values(ascending=False)

# 4. 시각화 (막대그래프)
plt.figure(figsize=(8, 5))
sns.barplot(x=correlations.values, y=correlations.index, palette="viridis")
plt.title("CACS와 변수 간 상관관계 (Pearson)")
plt.xlabel("상관계수")
plt.axvline(0, color="gray", linestyle="--")
plt.tight_layout()
plt.show()

# 5. 상관계수 수치 출력
print(correlations)
```
