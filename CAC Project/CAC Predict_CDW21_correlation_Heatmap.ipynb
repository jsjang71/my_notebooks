```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1.  
# -----------------------------
df = pd.read_excel("CACS_dataset_with_binary.xlsx")

#      
cols = ['age', 'SBP', 'DBP', 'BMI', 'TC', 'LDL', 'HDL', 'TG', 'CRP', 'NLR', 'eGFR']
data = df[cols].copy()

# -----------------------------
# 2.   (Spearman:  )
# -----------------------------
corr = data.corr(method='spearman')

# -----------------------------
# 3.  (/ )
# -----------------------------
sns.set(style="white", font_scale=1.2, rc={"figure.dpi":120})
plt.figure(figsize=(10,8))

heatmap = sns.heatmap(
    corr, annot=True, fmt=".2f", cmap="coolwarm", 
    square=True, cbar_kws={"shrink": .8},
    linewidths=0.6, annot_kws={"size":11}
)


plt.tight_layout()
plt.savefig("Figure_Correlation_Heatmap.png", dpi=600, bbox_inches="tight")
plt.show()
```
