```python
# 1. 라이브러리 불러오기
import pandas as pd
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 2. 데이터 불러오기
df = pd.read_excel("CACS_risk_score_output.xlsx")  # 파일 경로가 정확한지 확인

# 3. ROC 계산
fpr, tpr, _ = roc_curve(df['CACS_binary'], df['risk_score'])
auc = roc_auc_score(df['CACS_binary'], df['risk_score'])

# 4. ROC 그래프 그리기
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0, 1], [0, 1], 'k--')  # 대각선 기준선
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Risk Score ROC Curve")
plt.legend()
plt.grid()
plt.show()
```
