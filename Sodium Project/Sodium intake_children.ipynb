```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 (왼쪽 두 항목만)
labels = [
    "2010\n(age 6–18)",
    "2019\n(age 15–27)"
]
values = [3977.7, 3379.1]
x_pos = [0, 1]

# 그래프 생성
plt.figure(figsize=(6, 5))
bars = plt.bar(x_pos, values, color='#1f77b4', width=0.6)

# X축 레이블
plt.xticks(x_pos, labels, fontsize=10)

# 값 텍스트 표시
for x, val in zip(x_pos, values):
    plt.text(x, val + 50, f'{val:.1f}', ha='center', fontsize=10)

# 타이틀 및 축 설정
plt.title("Sodium intake change of Korean children", fontsize=14)
plt.ylabel("mg")
plt.ylim(2000, 5000)

plt.tight_layout()
plt.show()
import matplotlib.pyplot as plt
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 (왼쪽 두 항목만)
labels = [
    "2010\n(age 6–18)",
    "2019\n(age 15–27)"
]
values = [1911.4, 1626.4]
x_pos = [0, 1]

plt.figure(figsize=(6, 5))
bars = plt.bar(x_pos, values, color='#1f77b4', width=0.6)

# 값 표시
for x, val in zip(x_pos, values):
    plt.text(x, val + 30, f'{val:.1f}', ha='center', fontsize=10)

# x축 레이블 수동 지정
plt.xticks(x_pos, labels, fontsize=10)

# 제목 및 축 설정
plt.title("Sodium intake change of Korean children\nsodium/1000kcal", fontsize=14)
plt.ylabel("mg / 1000kcal")
plt.ylim(500, 2500)

plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import matplotlib as mpl

# 글꼴 설정
mpl.rc('font', family='Arial')

# 데이터 (왼쪽 두 항목만)
labels = [
    "2010\n(age 6–18)",
    "2019\n(age 15–27)"
]
values = [13.34, 26.22]
x_pos = [0, 1]

# 그래프 생성
plt.figure(figsize=(6, 5))
bars = plt.bar(x_pos, values, color='#e66b2b', width=0.6)

# 값 표시
for x, val in zip(x_pos, values):
    plt.text(x, val + 1, f'{val:.2f}', ha='center', fontsize=10)

# X축 레이블
plt.xticks(x_pos, labels, fontsize=10)

# 제목 및 축 설정
plt.title("% Korean children consuming less than\nsodium intake goal", fontsize=14)
plt.ylabel("%")
plt.ylim(0, 35)

plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import matplotlib as mpl

# 글꼴 설정
mpl.rc('font', family='Arial')

# 데이터
labels = ["2010\n(age 6–18)", "2019\n(age 15–27)"]
x_pos = [0, 1]

# 그래프 데이터
intake_mg = [3977.7, 3379.1]
intake_per_kcal = [1911.4, 1626.4]
percentage_below_goal = [13.34, 26.22]

# 색상
colors = ['#1f77b4', '#1f77b4']
colors2 = ['#1f77b4', '#1f77b4']
colors3 = ['#e66b2b', '#e66b2b']

# 전체 figure 설정
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.subplots_adjust(wspace=0.4)

# (1) Sodium intake (mg)
axes[0].bar(x_pos, intake_mg, color=colors, width=0.6)
axes[0].set_title("Sodium intake", fontsize=12)
axes[0].set_ylabel("mg")
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(labels, fontsize=10)
axes[0].set_ylim(2000, 5000)
for x, val in zip(x_pos, intake_mg):
    axes[0].text(x, val + 50, f'{val:.1f}', ha='center', fontsize=10)

# (2) Sodium intake / 1000 kcal
axes[1].bar(x_pos, intake_per_kcal, color=colors2, width=0.6)
axes[1].set_title("Sodium intake per 1000 kcal", fontsize=12)
axes[1].set_ylabel("mg / 1000kcal")
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(labels, fontsize=10)
axes[1].set_ylim(500, 2500)
for x, val in zip(x_pos, intake_per_kcal):
    axes[1].text(x, val + 30, f'{val:.1f}', ha='center', fontsize=10)

# (3) % below sodium intake goal
axes[2].bar(x_pos, percentage_below_goal, color=colors3, width=0.6)
axes[2].set_title("% below sodium intake goal", fontsize=12)
axes[2].set_ylabel("%")
axes[2].set_xticks(x_pos)
axes[2].set_xticklabels(labels, fontsize=10)
axes[2].set_ylim(0, 35)
for x, val in zip(x_pos, percentage_below_goal):
    axes[2].text(x, val + 1, f'{val:.2f}', ha='center', fontsize=10)

# 전체 제목
fig.suptitle("Sodium intake change of Korean children (2010 vs 2019)", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

import matplotlib.pyplot as plt
import matplotlib as mpl

# 글꼴 설정
mpl.rc('font', family='Arial')

# 데이터
labels = ["2010\n(age 6–18)", "2019\n(age 15–27)"]
x_pos = [0, 1]

# 그래프 데이터
intake_mg = [3977.7, 3379.1]
intake_per_kcal = [1911.4, 1626.4]
percentage_below_goal = [13.34, 26.22]

# 색상
colors1 = ['#1f77b4', '#1f77b4']       # 파랑: 첫 번째 그래프
colors2 = ['#f28e2b', '#f28e2b']       # 주황: 두 번째 그래프 (바뀐 색)
colors3 = ['#e66b2b', '#e66b2b']       # 진한 주황: 세 번째 그래프

# 전체 figure 설정
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.subplots_adjust(wspace=0.4)

# (a) Sodium intake (mg)
axes[0].bar(x_pos, intake_mg, color=colors1, width=0.6)
axes[0].set_title("Sodium intake", fontsize=12)
axes[0].set_ylabel("mg")
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(labels, fontsize=10)
axes[0].set_ylim(2000, 5000)
for x, val in zip(x_pos, intake_mg):
    axes[0].text(x, val + 50, f'{val:.1f}', ha='center', fontsize=10)
axes[0].text(-0.3, 4900, '(a)', fontsize=12, fontweight='bold')

# (b) Sodium intake / 1000 kcal (색상 변경됨)
axes[1].bar(x_pos, intake_per_kcal, color=colors2, width=0.6)
axes[1].set_title("Sodium intake per 1000 kcal", fontsize=12)
axes[1].set_ylabel("mg / 1000kcal")
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(labels, fontsize=10)
axes[1].set_ylim(500, 2500)
for x, val in zip(x_pos, intake_per_kcal):
    axes[1].text(x, val + 30, f'{val:.1f}', ha='center', fontsize=10)
axes[1].text(-0.3, 2400, '(b)', fontsize=12, fontweight='bold')

# (c) % below sodium intake goal
axes[2].bar(x_pos, percentage_below_goal, color=colors3, width=0.6)
axes[2].set_title("% below sodium intake goal", fontsize=12)
axes[2].set_ylabel("%")
axes[2].set_xticks(x_pos)
axes[2].set_xticklabels(labels, fontsize=10)
axes[2].set_ylim(0, 35)
for x, val in zip(x_pos, percentage_below_goal):
    axes[2].text(x, val + 1, f'{val:.2f}', ha='center', fontsize=10)
axes[2].text(-0.3, 33, '(c)', fontsize=12, fontweight='bold')

# 전체 제목
fig.suptitle("Sodium intake change of Korean children (2010 vs 2019)", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

import matplotlib.pyplot as plt
import matplotlib as mpl

# 글꼴 설정
mpl.rc('font', family='Arial')

# 데이터
labels = ["2010\n(age 6–18)", "2019\n(age 15–27)"]
x_pos = [0, 1]

# 그래프 데이터
intake_mg = [3977.7, 3379.1]
intake_per_kcal = [1911.4, 1626.4]
percentage_below_goal = [13.34, 26.22]

# 색상
colors1 = ['#1f77b4', '#1f77b4']
colors2 = ['#f28e2b', '#f28e2b']
colors3 = ['#e66b2b', '#e66b2b']

# 전체 figure 설정
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False)
fig.subplots_adjust(wspace=0.4)

# (a) Sodium intake (mg)
axes[0].bar(x_pos, intake_mg, color=colors1, width=0.6)
axes[0].set_title("Sodium intake", fontsize=12)
axes[0].set_ylabel("mg")
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(labels, fontsize=10)
axes[0].set_ylim(2000, 5000)
for x, val in zip(x_pos, intake_mg):
    axes[0].text(x, val + 50, f'{val:.1f}', ha='center', fontsize=10)
axes[0].text(-0.3, 4950, '(a)', fontsize=12, fontweight='bold')  # 1mm 아래로 조정 (~50 units)

# (b) Sodium intake per 1000 kcal
axes[1].bar(x_pos, intake_per_kcal, color=colors2, width=0.6)
axes[1].set_title("Sodium intake per 1000 kcal", fontsize=12)
axes[1].set_ylabel("mg / 1000kcal")
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(labels, fontsize=10)
axes[1].set_ylim(500, 2500)
for x, val in zip(x_pos, intake_per_kcal):
    axes[1].text(x, val + 30, f'{val:.1f}', ha='center', fontsize=10)
axes[1].text(-0.3, 5000, '(b)', fontsize=12, fontweight='bold')  # 기준 높이 유지

# (c) % below sodium intake goal
axes[2].bar(x_pos, percentage_below_goal, color=colors3, width=0.6)
axes[2].set_title("% below sodium intake goal", fontsize=12)
axes[2].set_ylabel("%")
axes[2].set_xticks(x_pos)
axes[2].set_xticklabels(labels, fontsize=10)
axes[2].set_ylim(0, 35)
for x, val in zip(x_pos, percentage_below_goal):
    axes[2].text(x, val + 1, f'{val:.2f}', ha='center', fontsize=10)
axes[2].text(-0.3, 33, '(c)', fontsize=12, fontweight='bold')

# 전체 제목
fig.suptitle("Sodium intake change of Korean children (2010 vs 2019)", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
