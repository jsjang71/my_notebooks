```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 생성
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'age>=1': [4831.1, 4789.3, 4583.1, 3862.1, 3754.7, 3889.6, 3338.5, 3337.9, 3274.1, 3243.1],
    'age 1~18': [3480.5, 3393.9, 3386.9, 2850.9, 2857.9, 2874.7, 2553.5, 2514.5, 2505.6, 2535.2],
    'age>=19': [5193.8, 5155.8, 4884.3, 4107.6, 3961.4, 4111.3, 3507.8, 3510.9, 3427.7, 3380.4]
})

# 누적 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['age>=1'], label='age>=1')
plt.bar(data['Year'], data['age 1~18'], bottom=data['age>=1'], label='age 1~18')
plt.bar(data['Year'], data['age>=19'], bottom=data['age>=1'] + data['age 1~18'], label='age>=19')

# 그래프 제목과 레이블 설정
plt.title('Sodium Intake Change 2010-2019')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(data['Year'])  # 모든 년도 표시
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=3, frameon=False)  # 범례와 Year 간 간격 좁힘
plt.show()
```
