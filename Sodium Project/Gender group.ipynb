```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 정의
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Male': [5666.2, 5648.0, 5280.2, 4508.2, 4370.3, 4619.7, 3948.4, 3886.5, 3866.8, 3811.1],
    'Female': [3991.3, 3929.7, 3884.6, 3215.1, 3138.7, 3160.2, 2725.3, 2786.7, 2678.6, 2673.3]
})

# 꺾은선 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Male'], marker='o', label='male', color='darkblue')
plt.plot(data['Year'], data['Female'], marker='o', label='female', color='darkorange')

# 주요 값에 텍스트 주석 추가
plt.annotate('5666.2', xy=(2010, 5666.2), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3991.3', xy=(2010, 3991.3), xytext=(-15, 5), textcoords='offset points')
plt.annotate('4619.7', xy=(2015, 4619.7), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3160.2', xy=(2015, 3160.2), xytext=(-15, 5), textcoords='offset points')

# 축, 제목, 범례 설정
plt.title('Sodium intake change 2010-2019')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(data['Year'])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=2, frameon=False)  # 범례와 Year 간 간격 좁힘
plt.grid(axis='y')  # 세로선 제거, 가로선만 표시

# 그래프 출력
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 정의
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Male': [5666.2, 5648.0, 5280.2, 4508.2, 4370.3, 4619.7, 3948.4, 3886.5, 3866.8, 3811.1],
    'Female': [3991.3, 3929.7, 3884.6, 3215.1, 3138.7, 3160.2, 2725.3, 2786.7, 2678.6, 2673.3]
})

# 누적 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['Male'], label='male', color='darkblue')
plt.bar(data['Year'], data['Female'], bottom=data['Male'], label='female', color='darkorange')

# 축, 제목, 범례 설정
plt.title('Sodium intake change 2010-2019')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(data['Year'])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=2, frameon=False)  # 범례 하단 배치, Year와 간격 좁힘
plt.grid(axis='y')  # 세로선 제거, 가로선만 표시

# 그래프 출력
plt.show()
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 정의
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Male': [5666.2, 5648.0, 5280.2, 4508.2, 4370.3, 4619.7, 3948.4, 3886.5, 3866.8, 3811.1],
    'Female': [3991.3, 3929.7, 3884.6, 3215.1, 3138.7, 3160.2, 2725.3, 2786.7, 2678.6, 2673.3]
})

# 산점도 및 추세선 생성
plt.figure(figsize=(10, 6))

# 남성 데이터
plt.scatter(data['Year'], data['Male'], color='darkblue', label='male')
male_fit = np.polyfit(data['Year'], data['Male'], 1)
plt.plot(data['Year'], np.polyval(male_fit, data['Year']), color='darkblue', linestyle='--')

# 여성 데이터
plt.scatter(data['Year'], data['Female'], color='darkorange', label='female')
female_fit = np.polyfit(data['Year'], data['Female'], 1)
plt.plot(data['Year'], np.polyval(female_fit, data['Year']), color='darkorange', linestyle='--')

# 축, 제목, 범례 설정
plt.title('Sodium intake change 2010-2019 (Scatter + Trendline)')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(data['Year'])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=2, frameon=False)
plt.grid(axis='y')

# 그래프 출력
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# 글씨체 설정
mpl.rc('font', family='Arial')

# 데이터 정의
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Male': [5666.2, 5648.0, 5280.2, 4508.2, 4370.3, 4619.7, 3948.4, 3886.5, 3866.8, 3811.1],
    'Female': [3991.3, 3929.7, 3884.6, 3215.1, 3138.7, 3160.2, 2725.3, 2786.7, 2678.6, 2673.3]
})

# 그룹형 막대그래프 생성
x = np.arange(len(data['Year']))  # X축 위치
width = 0.4  # 막대 너비

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, data['Male'], width=width, label='male', color='darkblue')
plt.bar(x + width/2, data['Female'], width=width, label='female', color='darkorange')

# 축, 제목, 범례 설정
plt.title('Sodium intake change 2010-2019 (Grouped Bar Chart)')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(x, data['Year'])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=2, frameon=False)
plt.grid(axis='y')

# 그래프 출력
plt.show()
```
