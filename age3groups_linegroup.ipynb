```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

#  
mpl.rc('font', family='Arial')

#  
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'age>=1': [4831.1, 4789.3, 4583.1, 3862.1, 3754.7, 3889.6, 3338.5, 3337.9, 3274.1, 3243.1],
    'age 1~18': [3480.5, 3393.9, 3386.9, 2850.9, 2857.9, 2874.7, 2553.5, 2514.5, 2505.6, 2535.2],
    'age>=19': [5193.8, 5155.8, 4884.3, 4107.6, 3961.4, 4111.3, 3507.8, 3510.9, 3427.7, 3380.4]
})

#  
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['age>=1'], marker='o', label='age>=1', color='blue')
plt.plot(data['Year'], data['age 1~18'], marker='o', label='age 1~18', color='orange')
plt.plot(data['Year'], data['age>=19'], marker='o', label='age>=19', color='green')

#     
plt.annotate('5193.8', xy=(2010, 5193.8), xytext=(-15, 5), textcoords='offset points')
plt.annotate('4831.1', xy=(2010, 4831.1), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3480.5', xy=(2010, 3480.5), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3961.4', xy=(2014, 3961.4), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3754.7', xy=(2014, 3754.7), xytext=(-15, 5), textcoords='offset points')
plt.annotate('2857.9', xy=(2014, 2857.9), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3380.4', xy=(2019, 3380.4), xytext=(-15, 5), textcoords='offset points')
plt.annotate('3243.1', xy=(2019, 3243.1), xytext=(-15, 5), textcoords='offset points')
plt.annotate('2535.2', xy=(2019, 2535.2), xytext=(-15, 5), textcoords='offset points')

#    
plt.title('Sodium intake change 2010-2019')
plt.xlabel('Year')
plt.ylabel('Sodium Intake (mg)')
plt.xticks(data['Year'])  #   
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, frameon=False)  #   , Year   
plt.grid(axis='y')  #  

#  
plt.show()
```
