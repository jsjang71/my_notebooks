```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

#  
mpl.rc('font', family='Arial')

#  
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Total': [11.75, 13.23, 14.04, 21.20, 21.25, 20.80, 26.20, 25.83, 25.84, 26.02],
    '2~18': [15.18, 15.68, 18.49, 25.32, 25.28, 25.06, 30.30, 30.95, 31.14, 31.00],
    '>=19': [10.87, 12.61, 12.98, 20.24, 20.37, 19.91, 25.36, 24.80, 24.82, 25.09]
})

#    
x = np.arange(3)  # Total, 2~18, >=19
bar_width = 0.08
colors = plt.cm.Blues(np.linspace(0.4, 1, 10))

plt.figure(figsize=(12, 6))

#   (bar) 
for i, year in enumerate(data['Year']):
    values = [data['Total'][i], data['2~18'][i], data['>=19'][i]]
    plt.bar(x + i * bar_width, values, width=bar_width, color=colors[i], label=str(year))

# , ,  
plt.xticks(x + bar_width * 4.5, ['Total', '2~18', '19'], fontsize=12)
plt.ylabel('%', fontsize=12)
plt.title('% Korean population with sodium intake <WHO goal', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

#  
mpl.rc('font', family='Arial')

#  
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Total': [11.75, 13.23, 14.04, 21.20, 21.25, 20.80, 26.20, 25.83, 25.84, 26.02],
    '2~18': [15.18, 15.68, 18.49, 25.32, 25.28, 25.06, 30.30, 30.95, 31.14, 31.00],
    '>=19': [10.87, 12.61, 12.98, 20.24, 20.37, 19.91, 25.36, 24.80, 24.82, 25.09]
})

#    
x = np.arange(3)  # Total, 2~18, >=19
bar_width = 0.06  #    
bar_gap = 0.01    #    
colors = plt.cm.Blues(np.linspace(0.4, 1, 10))

plt.figure(figsize=(12, 6))

#   (bar) 
for i, year in enumerate(data['Year']):
    values = [data['Total'][i], data['2~18'][i], data['>=19'][i]]
    plt.bar(x + i * (bar_width + bar_gap), values, width=bar_width, color=colors[i], label=str(year))

# , ,  
plt.xticks(x + ((bar_width + bar_gap) * 9) / 2, ['Total', '2~18', '19'], fontsize=12)
plt.ylabel('%', fontsize=12)
plt.title('% Korean population with sodium intake <WHO goal', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=5, frameon=False)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

#  
mpl.rc('font', family='Arial')

#  
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Total': [11.75, 13.23, 14.04, 21.20, 21.25, 20.80, 26.20, 25.83, 25.84, 26.02],
    '2~18': [15.18, 15.68, 18.49, 25.32, 25.28, 25.06, 30.30, 30.95, 31.14, 31.00],
    '>=19': [10.87, 12.61, 12.98, 20.24, 20.37, 19.91, 25.36, 24.80, 24.82, 25.09]
})

#    
x = np.arange(3)  # Total, 2~18, >=19
bar_width = 0.06  #    
bar_gap = 0.01    #    
colors = plt.cm.Blues(np.linspace(0.4, 1, 10))

plt.figure(figsize=(12, 6))

#   (bar) 
for i, year in enumerate(data['Year']):
    values = [data['Total'][i], data['2~18'][i], data['>=19'][i]]
    plt.bar(x + i * (bar_width + bar_gap), values, width=bar_width, color=colors[i], label=str(year))

# , ,  
plt.xticks(x + ((bar_width + bar_gap) * 9) / 2, ['Total', '2~18', '19'], fontsize=12)
plt.ylabel('%', fontsize=12)
plt.title('% Korean population with sodium intake <WHO goal', fontsize=14)
plt.legend(title='Year', bbox_to_anchor=(0.5, -0.13), loc='upper center', ncol=5, frameon=False)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
