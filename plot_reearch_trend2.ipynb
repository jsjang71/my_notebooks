```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#  
data = {
    "Publication Year": [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    "Monographs": [1, 2, 2, 0, 2, 1, 2, 7, 6, 2, 1, 2, 6],
    "Research Reports": [1, 0, 0, 1, 2, 1, 1, 1, 0, 0, 0, 0, 0],
    "Academic Articles": [0, 2, 3, 11, 5, 6, 10, 7, 7, 12, 10, 4, 10],
    "Dissertations": [2, 6, 1, 5, 1, 4, 4, 11, 17, 11, 11, 11, 9],
}

df = pd.DataFrame(data)

# Set figure size
plt.figure(figsize=(12, 6))
bar_width = 0.2
x_indexes = np.arange(len(df["Publication Year"]))

# Create the bar chart
bars1 = plt.bar(x_indexes - 1.5*bar_width, df["Monographs"], width=bar_width, label="Monographs", color='blue')
bars2 = plt.bar(x_indexes - 0.5*bar_width, df["Research Reports"], width=bar_width, label="Research Reports", color='red')
bars3 = plt.bar(x_indexes + 0.5*bar_width, df["Academic Articles"], width=bar_width, label="Academic Articles", color='green')
bars4 = plt.bar(x_indexes + 1.5*bar_width, df["Dissertations"], width=bar_width, label="Dissertations", color='purple')

# Adding labels above bars
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only label bars with non-zero height
            plt.text(bar.get_x() + bar.get_width()/2.0, height, str(int(height)),
                     ha='center', va='bottom', fontsize=9, fontweight='bold')

# Labels and title
plt.xlabel("Publication Year")
plt.ylabel("Number of Publications")
plt.title("Trends in Domestic Sodium Reduction Publications")
plt.xticks(ticks=x_indexes, labels=df["Publication Year"], rotation=45)
plt.legend()

# Save and display the chart
plt.tight_layout()
output_path = "/mnt/data/Sodium_Reduction_Publications_Annotated.png"
plt.savefig(output_path)
plt.show()

# Provide the download link
output_path
```
